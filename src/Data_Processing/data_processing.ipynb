## Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries ready")
```

## Step 2 — Load Raw Data

```python
df = pd.read_csv('../../data/raw/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1_8.csv')
print("Shape:", df.shape)
df.head()
```

## Step 3 — Basic Inspection

```python
# Data types and null check
print(df.dtypes)
print("\nMissing values:", df.isnull().sum().sum())
print("Duplicates    :", df.duplicated().sum())
```

```python
# Unique values in categorical columns
for col in df.select_dtypes('object').columns:
    print(f"{col}: {df[col].unique().tolist()}")
```

## Step 4 — Drop Identifier Column

```python
# EmpNumber is just an ID — not useful for modelling
df = df.drop(columns=['EmpNumber'])
print("Shape after drop:", df.shape)
```

## Step 5 — Encode Binary Columns

```python
# Yes/No → 1/0
df['OverTime']  = df['OverTime'].map({'Yes': 1, 'No': 0})
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
print("OverTime unique :", df['OverTime'].unique())
print("Attrition unique:", df['Attrition'].unique())
```

## Step 6 — Encode Ordinal Column

```python
# BusinessTravelFrequency has a natural order
df['BusinessTravelFrequency'] = df['BusinessTravelFrequency'].map({
    'Non-Travel'       : 0,
    'Travel_Rarely'    : 1,
    'Travel_Frequently': 2
})
print("BusinessTravelFrequency unique:", df['BusinessTravelFrequency'].unique())
```

## Step 7 — One-Hot Encode Nominal Columns

```python
# No natural order → create dummy columns
nominal = ['Gender', 'EducationBackground', 'MaritalStatus', 'EmpDepartment', 'EmpJobRole']
df = pd.get_dummies(df, columns=nominal)

# Convert True/False → 1/0
bool_cols = df.select_dtypes('bool').columns
df[bool_cols] = df[bool_cols].astype(int)

print("Shape after encoding:", df.shape)
print("Any non-numeric columns left?", df.select_dtypes('object').columns.tolist())
```

## Step 8 — Save Processed Data

```python
df.to_csv('../../data/processed/employee_performance_processed.csv', index=False)
print("Saved! Final shape:", df.shape)
```
